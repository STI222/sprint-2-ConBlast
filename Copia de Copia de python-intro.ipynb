```python
nombres = list(datos['Nombre'])

# Funci√≥n para calcular la similitud coseno entre dos vectores
def similitud_coseno(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    magnitude1 = np.linalg.norm(vector1)
    magnitude2 = np.linalg.norm(vector2)
    similitud = dot_product / (magnitude1 * magnitude2)
    return similitud


print("Seleccionar la primera persona:")
for i, nombre in enumerate(nombres):
    print(f"{i+1}. {nombre}")
persona1_index = nombres.index(input("Ingresa el nombre correspondiente: "))
persona1 = np.array(datos.iloc[persona1_index, 1:])


print("Seleccionar la segunda persona:")
for i, nombre in enumerate(nombres):
    print(f"{i+1}. {nombre}")
persona2_index = nombres.index(input("Ingresa el nombre correspondiente: "))
persona2 = np.array(datos.iloc[persona2_index, 1:])

similitud = similitud_coseno(persona1, persona2)

print(f"La similitud coseno de {nombres[persona1_index]} y {nombres[persona2_index]} es: {similitud:.4f}")
```
